```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
import time
import os
import math
%matplotlib inline
```

```python
def rescale_image(image, percent=75):
    width  = int(image.shape[1] * percent/ 100)
    height = int(image.shape[0] * percent/ 100)
    dim = (width, height)
    return cv2.resize(image, dim, interpolation =cv2.INTER_AREA)
```

## Get features of the 5 objects

```python
directory = './data/img'
i = 1
train_descriptors = dict()
train_images      = dict()
train_keypointss  = dict()
plt.figure(figsize=(10,10))

# Create an ORB object.
orb = cv2.ORB_create()

for item in os.listdir(directory):
    if os.path.isdir(directory + '/' + item):
        # load image
        image = cv2.imread(directory + '/' + item + '/' + item + '1.jpg')
        
        # lower resolution 10 folds
        image = rescale_image(image, percent=10)
        
        # convert the training image to RGB
        training_image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
        train_images[item] = training_image
        
        # convert the training image to gray scale
        training_gray  = cv2.cvtColor(training_image, cv2.COLOR_RGB2GRAY)
        
        # get orb features & store the descriptors in a dict
        train_keypoints, train_descriptor = orb.detectAndCompute(training_gray, None)
        train_descriptors[item] = train_descriptor
        train_keypointss[item]   = train_keypoints
        
        # plot keypoints
        keypoints_without_size = np.copy(training_image)
        plt.subplot(3,3,i)
        cv2.drawKeypoints(training_image, train_keypoints, keypoints_without_size, color = (0, 255, 0))
        plt.imshow(keypoints_without_size, cmap='gray')
        plt.title("Train keypoints With Size")
        
        i +=1
```

```python
def analyzeframe(frame, train_descriptors, bf):
    # get orb features 
    frame_keypoints, frame_descriptor = orb.detectAndCompute(frame, None)
    
    # compare the distance between the frame descriptor & the training descriptors
    distanceg = math.inf
    result    = ""
    matches_res = list()
    
    for item in train_descriptors:
        matches = bf.match(frame_descriptor, train_descriptors[item])
        matches = sorted(matches, key = lambda x : x.distance)

        # only keep good matches < 50
        #filter(lambda m: m.distance < 0, matches)
        
        # compute global distance
        distance = 0
        for m in matches:
            distance = m.distance
                
        if distance < distanceg:
            distanceg = distance
            result    = item
            matches_res = matches
            
    return result, matches_res, frame_keypoints
        
```

```python
frame = cv2.imread("./data/img/tasseg/tasseg3.jpg")
frame = rescale_image(frame, percent=10)
frame = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
frame_gray  = cv2.cvtColor(frame, cv2.COLOR_RGB2GRAY)

bf = cv2.BFMatcher(cv2.NORM_HAMMING, crossCheck = True)


result, matches, frame_keypoints = analyzeframe(frame_gray, train_descriptors, bf)

x = cv2.drawMatches(train_images[result], train_keypointss[result], frame_gray, frame_keypoints, matches, frame_gray, flags = 2)

#cv2.imshow('Frame',x)
print(result)
plt.imshow(x)
plt.show()
```

```python
cap = cv2.VideoCapture('./data/videos/video1_lowres.mp4')
bf = cv2.BFMatcher(cv2.NORM_HAMMING, crossCheck = True)

if (cap.isOpened()== False):
    print("Error opening video stream or file")

while(cap.isOpened()):
    ret, frame = cap.read()

    if ret == False:
        break
    else:
        frame  = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
        frame  = cv2.cvtColor(frame, cv2.COLOR_RGB2GRAY)
        result, matches, frame_keypoints = analyzeframe(frame, train_descriptors, bf)
        
        #x = cv2.drawKeypoints(frame, frame_keypoints, frame, color = (0, 255, 0))
        x = cv2.drawMatches(train_images[result], train_keypointss[result], frame, frame_keypoints, matches, frame, flags=cv2.DRAW_MATCHES_FLAGS_NOT_DRAW_SINGLE_POINTS)

        cv2.imshow('Frame',x)
        if cv2.waitKey(25) & 0xFF == ord('q'):
            break
            

    #if ret == True:
    #    cv2.imshow('Frame',frame)
    #    if cv2.waitKey(25) & 0xFF == ord('q'):
    #        break
    #else:
    #    break
    

        
cap.release()
cv2.destroyAllWindows()
```
